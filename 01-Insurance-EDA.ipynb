```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = data[['smoker', 'bmi', 'age']]
dataset
```

```python
# chart to display bmi and age of smokers and non-smokers

fig, ax = plt.subplots()

dataset[dataset['smoker'] == 'no'].plot.scatter(x='bmi', y='age', c='b', alpha=.25, ax=ax)
dataset[dataset['smoker'] == 'yes'].plot.scatter(x='bmi', y='age', c='r', alpha=.25, ax=ax)
```
